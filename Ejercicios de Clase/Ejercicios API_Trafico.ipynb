```python
import pandas as pd
from lxml import etree
import requests
url = 'https://informo.madrid.es/informo/tmadrid/pm.xml'
response = requests.get(url)
print(response)
df = pd.read_xml(response.content, xpath=".//pm")
df.head()
```

```python

import seaborn as sns

sns.displot(df['intensidadSat'], kind='kde')

sns.displot(df['ocupacion'], kind='hist')
```
