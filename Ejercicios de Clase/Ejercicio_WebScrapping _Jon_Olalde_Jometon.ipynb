```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
# from splinter import Browser
import numpy as np

pd.set_option('max_colwidth', 800)
```

```python
from fake_useragent import UserAgent
ua = UserAgent()
url = "https://www.elcorreo.com/temas/generales/accidentes-de-trafico.html"
headers = {'User-Agent': ua.random}
response = requests.get(url, headers=headers)
print(response)
```

```python
html = response.content
soup = bs(html, 'html.parser')
print(soup)
```

```python

noticias = []

for item in soup.select("article"):
    titulo_tag = item.find(['h2', 'h3'])    
    autor_tag = item.find('span', class_='v-op__t')
    autor = None
    if autor_tag:
        autor_a_tag = autor_tag.find('a')
        if autor_a_tag:
            autor = autor_a_tag.get('title')
    link_tag = item.find('a', href=True)    
    fecha_tag = item.find('p', class_='v-a-cro-t')
    hora_tag = fecha_tag.find('span') if fecha_tag else None

    if titulo_tag and link_tag:
        titulo = titulo_tag.get_text(strip=True)
        link = link_tag['href']
        
        if link.startswith('/'):
            link = 'https://www.elcorreo.com' + link 
        
        if fecha_tag:
            fecha = fecha_tag.get_text(strip=True).replace("\n", " ").strip()  
            hora = hora_tag.get_text(strip=True) if hora_tag else None
        else:
            fecha = None
            hora = None        
        noticias.append({
            'Título': titulo,
            'Autor': autor,
            'Enlace': link,
            'Fecha': fecha,
            'Hora': hora
        })

df = pd.DataFrame(noticias)
df.head(20)


```

```python
palabras_clave = ['accidente']

df_filtrado = df[df['Título'].str.lower().str.contains('|'.join(palabras_clave))]
df_filtrado.head(20)
```
