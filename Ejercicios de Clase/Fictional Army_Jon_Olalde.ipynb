### Step 1. Cread un dataframe con la información de abajo

```python
# Datos de nuestra armada
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, None, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, None, 3, None, 2, None, 2, None, None, 1, 2, None],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}

armada = pd.DataFrame(raw_data)

armada
```

### Step 2. Buscad valores faltantes

```python
valoresfaltantes = armada[armada.isnull().any(axis=1)]

valoresfaltantes
```

### Step 3. Eliminaremos aquellos que superen un umbral de 30% de datos faltantes

```python
armada_clen = armada.groupby('regiment').apply(lambda x: x.dropna(thresh=0.7))
armada_clen
```

    ### Step 4. Rellenad los datos faltantes con la media de la serie

```python
armada_clen['readiness'] = armada_clen['readiness'].fillna(armada_clen['readiness'].mean())
armada_clen['deaths'] = armada_clen['deaths'].fillna(armada_clen['deaths'].mean())
armada_clen
```

### Step 5. Marcad como índice la columna `origin`

```python
armada_clen = armada_clen.set_index('origin')
armada_clen
```
