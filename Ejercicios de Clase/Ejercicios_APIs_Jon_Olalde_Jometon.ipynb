```python
import requests
import pandas as pd
```

```python
url_incidents = "https://api.euskadi.eus/traffic/v1.0/incidences"

response = requests.get(url_incidents)
data_incidents = response.json()
df_incidents = pd.DataFrame(data_incidents)
df_incidents.head(40)
```

```python

url_cameras = "https://api.euskadi.eus/traffic/v1.0/cameras"

response = requests.get(url_cameras)
data_cameras = response.json()
print("Claves en cameras:", data_cameras.keys())


    
df_cameras = pd.DataFrame(data_cameras.get('cameras', []))
print("CÃ¡maras:", df_cameras.shape)
print(df_cameras.columns) 
df_cameras.head(40)



```

```python

import pyproj
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# Folium necesita coordenadas decimales al parecer y las que vienen en el dataset son UTM por lo que he tenido que buscar la manera de convertirlo
transformer = pyproj.Transformer.from_crs("EPSG:25830", "EPSG:4326", always_xy=True)

def convertir_utm_a_latlon(x, y):
    lon, lat = transformer.transform(x, y)
    return lat, lon
df_cameras[['lat', 'lon']] = df_cameras.apply(
    lambda row: pd.Series(convertir_utm_a_latlon(row['longitude'], row['latitude'])),
    axis=1
)

mapa_camaras = folium.Map(location=[43.2627, -2.93499], zoom_start=12) # La ubicacion que he seleccionado es Bilbao en este caso ya que de ahi son los elementos del dataset
marker_cluster = MarkerCluster().add_to(mapa_camaras)

for _, row in df_cameras.iterrows():
    if pd.notnull(row['lat']) and pd.notnull(row['lon']):
        popup_html = f"<b>{row['cameraName']}</b>"
        if pd.notnull(row['urlImage']):
            popup_html += f"<br><img src='{row['urlImage']}' width='200'>"
        
        folium.Marker(
            location=[row['lat'], row['lon']],
            popup=popup_html,
            icon=folium.Icon(color='blue', icon='camera', prefix='fa')
        ).add_to(marker_cluster)

mapa_camaras
```

```python
year = 2025
month = 5
day = 21

url_flow = f"https://api.euskadi.eus/traffic/v1.0/flows/byDate/{year}/{month}/{day}"

response = requests.get(url_flow)
data_flow = response.json()
print("Claves en data_flow:", data_flow.keys()) 

flows_list = data_flow.get('flows', [])
df_flow = pd.DataFrame(flows_list)
df_flow.head(40)

```

```python
year = 2025
month = 5
day = 21

url_incidences = f"https://api.euskadi.eus/traffic/v1.0/incidences/byDate/{year}/{month}/{day}"

response = requests.get(url_incidences)
data_incidences = response.json()
print("Claves en incidencias:", data_incidences.keys())
incidences_list = data_incidences.get('incidences', [])
df_incidences = pd.DataFrame(incidences_list)
df_incidences.head(40)
```

```python
print(df_incidences.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_incidences = df_incidences[df_incidences['incidenceType'].notna()]
plt.figure(figsize=(12, 6))
sns.countplot(
    data=df_incidences,
    y='incidenceType',
    order=df_incidences['incidenceType'].value_counts().index,
    palette="Set2"
)

plt.title("NÃºmero de incidencias por tipo (21 de mayo de 2025)")
plt.xlabel("NÃºmero de incidencia")
print(df_incidences['incidenceType'].value_counts())

```

```python


mapa_donosti = folium.Map(location=[43.3214, -1.9855], zoom_start=13)
marker_cluster = MarkerCluster().add_to(mapa_donosti)
colores = {
    tipo: color for tipo, color in zip(
        df_incidences['incidenceType'].unique(),
        ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightblue']
    )
}

for _, row in df_incidences.iterrows():
    tipo = row['incidenceType']
    popup_text = f"<b>{tipo}</b><br>{row['incidenceName']}<br>{row['cityTown']}"
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=6,
        color=colores.get(tipo, 'gray'),
        fill=True,
        fill_opacity=0.8,
        popup=popup_text
    ).add_to(marker_cluster)
mapa_donosti
```
